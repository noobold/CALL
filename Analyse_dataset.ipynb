## Fichier déchets

```python
benne_verre = pd.read_csv('CALL-data/Déchets.csv')
```

```python
perso_pd.analyseDataFrame(benne_verre)
```

```python
benne_verre = benne_verre.drop(['objectid'
                                ], axis=1)
```

```python
display(benne_verre[benne_verre.duplicated()])
```

Il y a réelement des bennes en double sur certain site(il y a 2 bennes) d'oú les doublons.<br>
Je vais supprimer les colonnes inutiles.<br>

```python
benne_verre = benne_verre.drop(['flux','created_user','created_date','last_edited_user','last_edited_date'
                                ,'identifiant','pédale','lon','lon2','lon3','lat','lat2','lat3'
                                ,'dimensions_des_plaques_de_consi','relevé_gps_code_1','relevé_gps_code_2'
                                ,'nombre_de_logements_individuels','nombre_de_logements_collectifs'
                                ,'surface___terre__ciment__herbe_','convention','date_fin_convention'
                                ,'informations','nom','adresse','marque','modèle','couleur','etat_borne'
                                ,'date_de_mise_en_service','appartiennent_à','aériennes__ba___enterrées__be__'
                                ,'X','Y'
                                ], axis=1)
```

```python
#objectif : une ligne par ville avec : le nombre de benne ,le pourcentage de benne enterré
#,le volume total,le pourcentage de trappe gros producteur, le pourcentage Cliiink
#,pourcentage avec des consigne a jour
benne_verre.replace({'Annay' : 'Annay-sous-Lens'
                   ,'Éleu-dit-Leauwette' : 'Eleu-dit-Leauwette'
                   ,'Estevelles' :'Estévelles'}
                   ,inplace=True)
```

```python
#transformer les Oui en 1
benne_verre['consignes_de_tri_à_jour'].fillna(0,inplace=True)
benne_verre['trappe_gros_producteurs'] = benne_verre['trappe_gros_producteurs'].apply(lambda x: 1 if x =='Oui' else 0)
benne_verre['cliiink'] = benne_verre['cliiink'].apply(lambda x: 1 if x =='Oui' else 0)
benne_verre['consignes_de_tri_à_jour'] = benne_verre['consignes_de_tri_à_jour'].apply(lambda x: 1 if x =='Oui' else 0)
```

```python
#faire une colonne nombre de bennes
benne_verre['nb_benne']=1
```

```python
#regrouper les communes et faire la somme
benne_verre = benne_verre.groupby('commune').sum()
benne_verre.reset_index(inplace=True)
```

```python
fichier_final_dechet = benne_verre.merge(population,left_on='commune',right_on='Libelle',how='outer')
fichier_final_dechet.drop(['Libelle'],axis=1,inplace=True)
```

```python
fichier_final_dechet.to_csv('donnee/benne_verre.csv',index=False)
```
